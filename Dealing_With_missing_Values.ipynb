**Filling the Missing Values - Imputation**

**1) Numerical data - use mean**

```python
ddf2 = df.copy()
```

```python
ddf2.head()
```

```python
mean_values = ddf2["reviews_per_month"].mean()

ddf2["reviews_per_month"].fillna(mean_values, inplace = True)
```

```python
ddf2.head()
```

```python
ddf2.isnull().sum()
```

**Filling Categorical data**

```python
ddf2.nunique()
```

```python
ddf2["last_review"].value_counts()
```

```python
ddf2["last_review"].fillna(ddf2["last_review"].value_counts().index[0], inplace = True)
```

```python
ddf2.isnull().sum()
```

```python
ddf3 = df.copy()
```

```python
ddf3.isna().sum()
```

```python
ddf3["last_review"].fillna("Not Reiviewed", inplace = True)
```

```python
ddf3.sample(20)
```
